## Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
from matplotlib import style
```

```python
style.use('seaborn')
```

## Buscando dados

```python
pytrend = TrendReq(hl='en_US', tz=360)
```

## Palavra-chave

```python
lista = ['educação financeira']
```

```python
pytrend.build_payload(lista, cat=0, timeframe='2019-01-01 2021-10-01', geo='BR', gprop='youtube')
```

```python
dados = pytrend.interest_over_time()
dados.head()
```

```python
plt.figure(figsize=(12, 6));
plt.title('Pesquisas no Youtube por educação financeira', fontsize=15);
plt.plot(dados['educação financeira'], color='blue', label='Ed. financeira pré-pandemia');
plt.plot(dados['educação financeira'][dados.index.year >= 2020], color='red', label='Ed. financeira durante a pandemia');
plt.xlabel('Data', fontsize=13);
plt.ylabel('Freq. de pesquisa', fontsize=13);
plt.text(np.datetime64('2020-12-01'), 40, 'By: Observatório de Finanças - UFC\nGoogleTrends', fontsize=10);
plt.legend(loc=2);
```
